```python
import pandas as pd
import numpy as np
import anal_preprocess as ap
import seaborn as sns
import matplotlib.pyplot as plt
import get_numcatg_data as gnca
data = pd.read_csv("./insurance.csv")
data.head()
```

```python
import scipy.stats as stats
print('age shapiro value : ',stats.shapiro(data['age']))
print('charges shapiro value : ', stats.shapiro(data['charges']))
```

```python
from sklearn.linear_model import LinearRegression
x = np.array(data['age'])
y = np.array(data['charges'])
x = x.reshape(1338, 1)
# .reshape()으로 차원 맞추기
y = y.reshape(1338, 1)
lr = LinearRegression()
lr.fit(x,y)
```

```python
print('선형 회귀 모델 결과')
print('절편', lr.intercept_, '계수', lr.coef_)
```

```python
print(lr.score(x,y)) # 결정계수 r2가 작을 수록 선형관계가 작다는 뜻
```

```python
x_new=[[19],[64]]
y_hat = lr.predict(x_new)
print(y_hat)
```
